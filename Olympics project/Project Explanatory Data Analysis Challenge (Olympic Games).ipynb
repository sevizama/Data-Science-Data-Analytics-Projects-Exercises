## Hipothesis 1: there is no relationship between Total Medals and Population

```python
import scipy.stats as stats
```

```python
#Pvalue is very close to 0.
stats.spearmanr(medals_per_country['Total'], medals_per_country['Population'], nan_policy='omit')
```

```python
r, p_value = stats.spearmanr(medals_per_country['Total'], medals_per_country['Population'], nan_policy='omit')
```

```python
r
```

```python
round(p_value)
```

**En realidad, si el valor p es cercano a 0 en un análisis de regresión entre dos variables (en este caso, Population y Number of Medals), eso generalmente implica que hay una relación significativa entre las dos variables. En estadísticas, un valor p bajo generalmente significa que puedes rechazar la hipótesis nula, que en este caso sería que no hay relación entre la población y el número de medallas.

**Por lo tanto, si obtienes un valor p cercano a 0, esto sugiere que hay evidencia estadística para afirmar que existe una relación significativa entre la población y el número de medallas. En otras palabras, la población podría estar relacionada con el número de medallas en el contexto de tu análisis.

**Si deseas afirmar que no hay relación entre la población y el número de medallas, querrías obtener un valor p alto (generalmente cercano a 1), lo que indicaría que no hay evidencia estadística para respaldar la relación entre estas dos variables.

### P_VALUE NOS SALE CERCANO A 0, ESO IMPLICA QUE HAY RELACION ENTRE LAS VARIABLES POR LO QUE RECHAZAMOS LA HIPOTESIS NULA, ES DECIR LA QUE PLANTEAMOS.

### RECHAZAMOS LO SIGUIENTE: NO HAY RELACION ENTRE POPULATION Y Nº DE MEDALLAS

## Aggregating and Ranking

```python
sum_vs_win = pd.crosstab(olympics['Country_Name'],olympics['Edition'])
sum_vs_win
```

```python
man_vs_woman = pd.crosstab(olympics['Country_Name'],olympics['Gender'])
man_vs_woman
```

```python
medal_by_cats = pd.concat([sum_vs_win,man_vs_woman],axis = 1)
medal_by_cats
```

```python
medal_by_cats['Total_by_Edition'] = medal_by_cats['Summer'] + medal_by_cats['Winter']
```

```python
medal_by_cats
```

```python
medal_by_cats.sort_values(by='Total_by_Edition',ascending=False, inplace = True)
```

```python
medal_by_cats
```

```python
#If we see that the US has more medals in summer than in winter we could think they are less succesfull in winter but that is False
#There are less games in winter so to make a fair comparisson we will use this method
ranks  = medal_by_cats.rank(method='average',ascending = False)
ranks
```

```python
top_50 = ranks.head(50)
```

Create the following __Seaborn Heatmap__ with __Medal Ranks__ for Top 50 Countries (Total Medals, Summer Games Medals, Winter Games Medals, Men, Women).

```python
#We can see that UK man are more succesfull than UK woman and also that that UK is a summer sport country.
plt.figure(figsize = (50, 10))
sns.heatmap(top_50.T,cmap='RdYlGn_r',annot=True,fmt='2.0f')
plt.show()
```
